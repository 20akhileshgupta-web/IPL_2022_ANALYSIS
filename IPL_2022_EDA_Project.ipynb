# UNIVARIATE ANALYSIS

## FIRST INNING SCORE ANALYSIS

```python
plt.figure(figsize=(10,6), facecolor='lightgrey')

sns.histplot(
    df['first_ings_score'],
    bins=15,
    kde=True,
    color='royalblue',
    edgecolor='black',
    linewidth=1.2
)

plt.title('First Innings Score Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(alpha=0.3, linestyle='--')

plt.show()
```

## SECOND INNING SCORE ANALYSIS

```python
plt.figure(figsize=(10,6), facecolor='whitesmoke')

sns.histplot(
    df['second_ings_score'],
    bins=15,
    kde=True,
    color='steelblue',
    edgecolor='black',
    linewidth=1.2
)

plt.title('Second Innings Score Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(alpha=0.3, linestyle='--')

plt.show()
```

# TEAM WINNING ALAYSIS

```python
match_wins = df['match_winner'].value_counts() 
match_wins
```

```python
sns.barplot(x=match_wins.values, y=match_wins.index, palette='viridis')
plt.title("HIGHEST_MATCH_WINNERS")
plt.show()
```

# TOSS ANALYSIS

```python
toss_trends = df['toss_decision'].value_counts()
toss_trends
```

```python
plt.figure(figsize=(5, 4), facecolor='lightyellow')

plt.pie(
    toss_trends,
    labels=toss_trends.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05, 0.05],      # Slightly separate slices
    shadow=True,
    textprops={'fontsize': 14, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'black', 'linewidth': 2}
)

# Title
plt.title(
    "Toss Decision Distribution",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.tight_layout()
plt.show()
```

## FIRST INNING SCORE VS MATCH RESULT ANALYSIS

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.boxplot(
    data=df,
    x='won_by',
    y='first_ings_score',
    palette='coolwarm'
)

sns.stripplot(
    data=df,
    x='won_by',
    y='first_ings_score',
    color='black',
    alpha=0.4,
    jitter=True
)

plt.title("Score Distribution by Winning Margin", fontsize=18, fontweight='bold')
plt.show()
```

# CORRELATION ANALYSIS

```python
# selecting all the columns including numeric columns 
num_df = df.select_dtypes(include=np.number)
corr = num_df.corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()

# strong positive correlation → close to +1
# strong negative → close to -1
```
